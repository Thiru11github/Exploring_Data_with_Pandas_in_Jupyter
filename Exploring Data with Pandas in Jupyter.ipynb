```python
df = pd.read_csv(r"C:\Users\thiru\Downloads\weather_data (1).csv",parse_dates= ['day'])
df.set_index('day',inplace=True)
df
```

```python
#Fill all NaN with one specific value

new_df= df.fillna(0)
new_df
```

```python
#Fill na using column names and dict

new_df = df.fillna({
        'temperature': 0,
        'windspeed': 0,
        'event': 'No Event'
    })
new_df
```

```python
#ffill - forward fill
#it works from top to bottom
#Example: If you have [1, NaN, NaN, 2],ffill will fill the missing values with 1, resulting in [1, 1, 1, 2].
new_df = df.fillna(method = 'ffill')
new_df
```

```python
#backward fill 
#it works from bottom to top 
new_df = df.fillna(method='bfill')
new_df
```

```python
new_df = df.fillna(method='bfill',limit=1)
new_df
```

```python
new_df = df.fillna(method='bfill',axis = 'columns')
new_df
```

```python
#interpolate
new_df = df.interpolate()
new_df
```

```python
#axis=0 refers to operations along the rows (vertically).
#axis=1 refers to operations along the columns (horizontally).
```

```python
#dropna
new_df = df.dropna()
new_df
```

```python
new_df = df.dropna(how='all')
new_df
```

```python
#thresh-specify the minimum number of non-null values required for a row or column to be kept.
new_df = df.dropna(thresh=2)
new_df
```

```python
#merge DataFrames
```

```python
import pandas as pd
```

```python
df1 = pd.DataFrame({
    'city' : ['new york','chicago','orlando','baltimore'],
    'temperature' : [21,14,35,32]
})
df1
```

```python
df2 = pd.DataFrame({
    'city': ['chicago','new york','san francisco'],
    'humidity' : [65,68,71]
})
df2
```

```python
df = pd.merge(df1,df2,on='city')
df

#retrieve matching rows from table 1 and table 2 , kinda inner join
```

```python
df = pd.merge(df1,df2,on='city',how='outer',indicator = True)
df

#retrieve all data from table , kinda outer join
```

```python
df = pd.merge(df1,df2,on='city',how = 'left')
df

#retrieve all data from left table
```

```python
df = pd.merge(df1,df2,on='city',how = 'right')
df

#retrieve all data from right table
```

```python
#suffixes
df1 = pd.DataFrame({
    'city': ['new york','chicago','orlandp','batlimore'],
    'temperature' : [21,14,35,38],
    'humidity' : [65,68,71,75]
})
df1
```

```python
#suffixes
df2 = pd.DataFrame({
    'city': ['chicago','new york','san diego'],
    'temperature' : [21,14,35],
    'humidity' : [65,68,71]
})
df2
```

```python
df3 = pd.merge(df1,df2,on='city')
df3
```

```python
df3 = pd.merge(df1,df2,on='city',suffixes = ('_left','_right'))
df3
```
